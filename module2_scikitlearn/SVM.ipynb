```python
#support vector machine

from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
x, y = make_blobs(n_samples=30,centers=2, n_features=2, random_state=0, cluster_std=0.5)
plt.scatter(x[:,0],x[:,1], c=y)
plt.show()
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# y = (0.21459589*x1) +  (-1.28877637*x2) + 3.15611447
model.decision_function(x)
```

```python
import numpy as np
xin = np.arange(-1,4)
yin = np.arange(-1,7)
# print(xin,yin)
Yin, Xin = np.meshgrid(yin,xin)
# print(Yin,Xin)
xy = np.vstack([Xin.ravel(),Yin.ravel()]).T
# print(xy)
p = model.decision_function(xy).reshape(Xin.shape)

p
```

```python
plt.contour(Xin,Yin,p, levels=[-1, 0, 1], linestyles=['--', '-', '--'], colors='k')
plt.scatter(x[:,0],x[:,1], c=y)

plt.scatter(model.support_vectors_[:,0],model.support_vectors_[:,1], s=70, facecolors='none')
plt.show()
```
